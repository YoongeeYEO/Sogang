```python
data.columns = ['타임스탬프','성별','연령','문구','새로움1','새로움2','정확성1','정확성2','사회적 실재감1','사회적 실재감2','자기참조1','자기참조2','만족도1','만족도2','외식빈도']
```

```python
data['새로움'] = data['새로움1']+data['새로움2']
```

```python
data['정확성'] = data['정확성1']+data['정확성2']
```

```python
data['사회적실재감'] = data['사회적 실재감1'] + data['사회적 실재감2']
```

```python
data['자기참조'] = data['자기참조1']+data['자기참조2']
```

```python
data['만족도'] = data['만족도1']+data['만족도2']
```

```python
target = data[['성별','연령','외식빈도','정확성','새로움','자기참조','사회적실재감','만족도']]
```

```python
target[['연령','자기참조']].groupby('연령').describe()
```

```python
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm 
```

```python
model = ols('만족도 ~  C(자기참조)* C(연령)', target).fit()
```

**비균형설계 자료의 경우 typ=3 옵션을 추가해 준다.**

```python
anova_lm(model, typ=3)
```
